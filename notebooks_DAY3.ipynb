```python
#comparison operators - all the operators used here will be used for comparing

== for equality 
!= not equality 
> greater than
< less than
>= greater or equals to 
<= less than or equals to 

```

```python
x,y = 5,10
print(x == y)
print(x != y)
print(y > x)
print(x < y)
print(x >= y)
print(x <= y)
```

```python
# logical operators , used to combine conditional statements 

and - returns true if both statements are true
or - returns true if one of the statement is true 
not - reverse the result, returns false if the result is true 
```

```python
x = 5

print(x > 0 and x < 10)
```

```python
print(x > 6 and x < 10)
```

```python
print(x > 6 or x < 10)
```

```python
print(not(x > 6 or x < 10))
```

```python
print(not(x > 0 and x < 10))
```

```python
# if - elif - else

a = int(input("Enter the value of a: "))
b = int(input("Enter the value of b:  "))

if a > b: 
    print("a is greater")
elif b > a:
    print("b is greater")
else:
    print("a is equal to b")
```
